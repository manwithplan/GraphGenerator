First we filter the job titles from the data

```python
titles = ['minister', 'governor', 'president', 'manager', 'director', 'sir', 'secretary', 'sheikh', 'representative','speaker','representative','deputy', 'ceo' ]
```

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_lg")
entities = nlp.get_pipe('ner').labels
```

```python
matcher = Matcher(nlp.vocab)
# Add match ID "HelloWorld" with no callback and one pattern
pattern = [
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'},
    {'ORTH': {'IN': titles}},
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'}
]
matcher.add("PERSON", [pattern])
```

```python
import random

ids = list(articles_cleaned.keys())

random_id = random.sample(ids, 1)

random_article = articles_cleaned[random_id[0]]
```

```python
doc = nlp(random_article)

matches = matcher(doc)

spans = []

for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    spans.append(span)

final_match = spacy.util.filter_spans(spans)
print(final_match)
```

```python
for token in doc[start - 5:end + 5]:
    print(token.text, token.pos_, token.dep_)
```

```python
pattern_person = [
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'},
    {"IS_PUNCT": True, "OP": "?"},
    {'ORTH': {'IN': titles}},
    {"IS_PUNCT": True, "OP": "?"},
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'}
]
```

```python
pattern_company = [
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'},
    {'POS': 'CCONJ'},
    {'ORTH': {'IN': formatted_tickers}},
    {'POS': {'IN': ['PROPN', 'NOUN']}, 'OP': '*'}
]
```

```python
pattern_commodity = [
    {'ORTH': {'IN': commodities_list}}
]
```

```python
from spacy.lang.en import English

nlp = spacy.load('en_core_web_lg')

# Construction from class
from spacy.pipeline import EntityRuler
ruler = EntityRuler(nlp)

ruler = nlp.add_pipe("entity_ruler", before="ner")

nlp.pipe_names
```

```python
patterns = [
    {"label": "PERSON", "pattern": pattern_person, "id": "Person of Interest"},
    {"label": "COMPANY", "pattern": pattern_company, "id": "Company"},
    {"label": "COMMODITY", "pattern": pattern_commodity, "id": "Comodity"}
]
```

```python
ruler.add_patterns(patterns)
```
